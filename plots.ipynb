```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_scores = np.load("test_scores.npy")
```

```python
plt.plot(
        test_scores[:, 0],
        test_scores[:, 1]
    
)


plt.xlabel("NOISE")
plt.ylabel("TEST SCORE")
plt.title("TEST LOSS v/s NOISE")
```

```python
train_params_0 = np.load("results/train_meta_param.npy")
train_params_1 = np.load("results/train_meta_param_1.npy")
train_params_2 = np.load("results/train_meta_param_2.npy")
train_params_3 = np.load("results/train_meta_param_3.npy")
```

```python
train_loss = np.hstack(
    [
        train_params_0[:, 0],
        train_params_1[:, 0],
        train_params_2[:, 0],
        train_params_3[:, 0]
    ]
)

test_loss = np.hstack(
    [
        train_params_0[:, 1],
        train_params_1[:, 1],
        train_params_2[:, 1],
        train_params_3[:, 1]
    ]
)

iou = np.hstack(
    [
        train_params_0[:, 2],
        train_params_1[:, 2],
        train_params_2[:, 2],
        train_params_3[:, 2]
    ]
)
```

```python
min_max = lambda a: (a - np.min(a))/(np.max(a) - np.min(a))

plt.plot(
        min_max(train_loss)
)
plt.plot(
        min_max(test_loss)
)
plt.plot(
        min_max(iou)
)


plt.xlabel("EPOCHS")
plt.title("TRAIN, TEST LOSS, IOU v/s EPOCHS")
```

```python
train_denoise = np.load("results/train_meta_denoiser.npy")
```

```python
for idx, i in enumerate(zip(train_loss, test_loss, iou)):
    print("Epoch {}:\t{}\t{}\t{}".format(idx, i[0], i[1], i[2]))
```

```python

plt.plot(
        min_max(train_denoise[:, 0])
)
plt.plot(
        min_max(train_denoise[:, 1])
)

plt.xlabel("EPOCHS")
plt.title("TRAIN, TEST LOSS v/s EPOCHS")
```
